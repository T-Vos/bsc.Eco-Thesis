```python
df_filtered = df[ df["contract_type"] == "ERC721"]
df_filtered = df_filtered.drop(["block_hash",
                                "block_number", 
                                "operator",
                                "verified",
                                "transaction_type",
                                "transaction_index",
                                "log_index",
                                "contract_type",
                                "token_id",
                                "from_address",
                                "to_address",
                                "amount"
                                ], axis=1)
```

```python
def shortenInt(value):
    value = str(value)
    if len(value) > 7 : value = value[:-7]
    return int(value)
```

```python
df_filtered["value_cutDown"] = df_filtered["value"].map(lambda x : shortenInt(x))
df_filtered['value_cutDown'] = pd.to_numeric(df_filtered['value_cutDown'])
df_filtered['transactionEther'] = df_filtered["value_cutDown"] / 100000000000
```

```python
df_filtered = df_filtered.drop(["value_cutDown",'value'], axis=1)
```

```python
df_filtered['block_timestamp'] = pd.to_datetime(df_filtered['block_timestamp'])
df_filtered['block_timestamp_day'] = df_filtered['block_timestamp'].dt.floor('d')
# df_filtered.sort_values("block_timestamp")
```

```python
import numpy as np

df_grouped = df_filtered.groupby(["transaction_hash"])

grouped = []

for (name, groupobject) in df_grouped.__iter__():
    price =  np.array(groupobject["transactionEther"])[0]
    timeStamp =  np.array(groupobject["block_timestamp"])[0]
    timeStamp_day =  np.array(groupobject["block_timestamp_day"])[0]
    nft = np.array(groupobject["token_address"])[0]
    amount = len(groupobject)
    grouped.append({
        "timeStamp":timeStamp,
        "timeStamp-day":timeStamp_day,
        "price":price,
        "quant": amount,
        "nft": nft,
        "transactie":name
    })

# trade price 
    # mean
    # high
    # low
    # difference previous day
    # volatility
# trade volume
```
