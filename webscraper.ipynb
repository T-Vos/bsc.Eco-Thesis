```python
import pandas as pd
def loadData(rows = 100) :
    return pd.read_csv("./data/Data_API.csv", nrows=rows)
```

```python
from webscraper.etherscan import webScraperPrice
df = loadData(100)
df["transactionPriceAtMoment"], df["transActionPriceNow"] = zip(*df['Transaction_hash'].map(webScraperPrice))
```

```python
df["transactionPriceAtMoment"] = df["transactionPriceAtMoment"].apply(lambda x: x.replace('(', '').replace(')','').replace('$',''))
df["transActionPriceNow"] = df["transActionPriceNow"].apply(lambda x: x.replace('(', '').replace(')','').replace('$',''))
```

```python
df["transactionPriceAtMoment"] =pd.to_numeric(df["transactionPriceAtMoment"])
df["transActionPriceNow"] = pd.to_numeric(df["transActionPriceNow"])
```
